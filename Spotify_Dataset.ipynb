```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path ='/content/drive/My Drive/Kaggle_spotify_dataset.csv'
spotify_df = pd.read_csv(path)
spotify_df.head()
```

```python
spotify_df.shape
```

```python
spotify_clean= spotify_df.drop(columns =['track_id','Unnamed: 0', 'time_signature', 'mode'], axis ='columns')
```

```python
spotify_clean.head()
```

```python
spotify_clean.drop_duplicates(subset=None, keep='first',inplace=False)
```

```python
spotify_clean.describe()
```

```python
spotify_clean.describe(include = ['object'])
```

```python
spotify_clean.info()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='explicit', data=spotify_clean, saturation=5, palette='crest')
sns.despine()
plt.title("Explicit")
```

```python
plt.figure(figsize=(10, 6))
bar1 = sns.barplot(x="explicit",  y="popularity", data=spotify_clean, palette='crest')
sns.despine()
plt.title("Popularity vs Explicit")
```

```python
plt.figure(figsize=(10, 6))
bar2 = sns.barplot(x="key",  y="popularity", data=spotify_clean, palette='crest')
sns.despine()
plt.title("Populatity by Key")
```

```python
spotify_clean.groupby('artists')['popularity'].max().reset_index().head(50)
```

```python
spotify_clean.groupby('artists')['popularity'].mean().reset_index().head(50).round(2)
```

```python
plt.figure(figsize=(10, 6))
bar2 = sns.barplot(x="explicit",  y="speechiness", data=spotify_clean, palette='crest')
sns.despine()
plt.title("Explicit Vs by Speechiness")
```

```python
plt.figure(figsize=(10, 6))
plt.title("Popularity Vs Key Vs Explicit")

sns.lineplot(x='key', y='popularity', data=spotify_clean, hue='explicit', errorbar=None, estimator=np.mean, palette='crest')

plt.xlabel("Key")
plt.ylabel('Popularity')
```

```python
plt.figure(figsize=(14, 6))
df_cmp = spotify_clean[['popularity', 'explicit', 'energy', 'key', 'speechiness', 'valence', 'tempo', 'danceability', 'duration_ms', 'loudness', 'acousticness', 'liveness', 'instrumentalness']]
correlation_matrix = df_cmp.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='GnBu', fmt='.2f', linewidths=0.75)
plt.title('Correlation Heatmap')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=spotify_clean['valence'], y=spotify_clean['danceability'])
plt.title('Valence vs Danceability')
sns.despine()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=spotify_clean['energy'], y=spotify_clean['loudness'])
plt.title('Energy vs Loudness')
sns.despine()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=spotify_clean['acousticness'], y=spotify_clean['energy'])
plt.title('Acousticness vs Energy')
sns.despine()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=spotify_clean['loudness'], y=spotify_clean['energy'])
plt.title('Loudness vs Energy')
sns.despine()
```
